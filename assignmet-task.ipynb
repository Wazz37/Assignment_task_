# Linear Regression Assignment

## *Question 1. Calculate the linear regression of patient's age and average maximum heart rate in beats per minute*

## Importing Libraries

```python
import numpy as np
import pandas as pd
```

### Reading Data

```python
df = pd.read_csv('../input/assignmet/Assignment_1 - Sheet2.csv')
df.head()
```

### Creating Dependent and Independent features

```python
x = df.iloc[:,1].values.reshape(-1,1)
y = df.iloc[:,-1].values.reshape(-1,1)
```

## Visualizaitons

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Checking for Outliers
### Boxplot

```python
sns.boxplot(x=df['y'],data=df)
```

No outliers are present

### Satter Plot

```python
plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Heart beat rate per minute')
plt.title('Age Vs Heart beat rate per minute')
```

## Model

### Importing regression model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

### Fitting model

```python
lr.fit(x,y)
```

```python
x_pred = lr.predict(x)
```

```python
lr.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,x_pred,color='red')
plt.title('Age vs Avg heartbeat per minute')
plt.xlabel('Age')
plt.ylabel('Avg hearbeat per minute')
```

```python
lr.coef_[0][0]
```

```python
lr.intercept_[0]
```
